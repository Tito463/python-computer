```python
import numpy as np

A=np.array([10,20,30,40,50])
B=np.array([1,2,3,4,5])
add=np.add(A,B)
print(add)
sub=np.subtract(A,B)
print(sub)
mul=np.multiply(A,B)
print(mul)
div=np.divide(A,B)
print(div)

mean_A=np.mean(A)
print(mean_A)
max_A=np.max(A)
print(max_A)
min_A=np.min(A)
print(min_A)
dot_AB=np.dot(A,B)
print(dot_AB)

reshaped_A=np.reshape(A,(5,1))
print(reshaped_A)
```
